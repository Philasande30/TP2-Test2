```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load the dataset
df = pd.read_csv("C:/Users/MS Mthethwa/Desktop/Nana/City Development Index.csv.xls")
```

```python
# Basic Infographic Charts
# 1. Bar chart showing the distribution of countries

# Limiting to the top 5 countries for better visualization
country_counts = df['Country'].value_counts().head(5)

# Setting up the style
sns.set_style("whitegrid")

# Creating the bar plot
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=country_counts.index, y=country_counts.values, palette="viridis")

# Adding annotations
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', fontsize=12, color='black', xytext=(0, 5),
                textcoords='offset points')

# Adding titles and labels
plt.title('Top 5 Countries by Frequency', fontsize=16)
plt.xlabel('Country', fontsize=14)
plt.ylabel('Count', fontsize=14)
plt.xticks(rotation=45, fontsize=12)
plt.yticks(fontsize=12)

# Displaying the plot
plt.tight_layout()
plt.show()
```

```python
# 2. Pie chart showing the distribution of regions
region_counts = df['Region'].value_counts()
plt.figure(figsize=(8, 8))
plt.pie(region_counts, labels=region_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Regions')
plt.axis('equal')
plt.show()
```

```python
# 3. Boxplot to visualize the distribution of CDI (Community Development Index) by region

plt.figure(figsize=(10, 6))
sns.boxplot(x='Region', y='CDI', data=df)
plt.title('Distribution of CDI by Region')
plt.xlabel('Region')
plt.ylabel('CDI')
plt.xticks(rotation=45)
plt.show()
```

```python
# Advanced Charts
# 4. Violin plot to visualize the distribution of Health
plt.figure(figsize=(8, 6))
sns.violinplot(y='Health', data=df)
plt.title('Violin Plot: Distribution of Health')
plt.ylabel('Health')
plt.show()
```

```python
# 5. Pair plot to explore relationships between numerical variables
sns.pairplot(df[['CDI', 'Infrastructure', 'Waste', 'Health', 'Education']])
plt.suptitle('Pair Plot of Numerical Variables', y=1.02)
plt.show()
```

```python
# Remove 'Unnamed' columns and non-numeric columns
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Calculate summary statistics
summary_stats = numeric_df.describe()

# Extract mean, median, and standard deviation
mean_values = numeric_df.mean()
median_values = numeric_df.median()
std_values = numeric_df.std()

# Print summary statistics with mean, median, and standard deviation
print("Summary Statistics with Mean, Median, and Standard Deviation:")
print(summary_stats)
```
